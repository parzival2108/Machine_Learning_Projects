```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

### The data is a subsampled version of the MNIST dataset. It contains images for 10 digits (10 classes). The dataset contains 6,000 samples. The images from the data set have the size 28 x 28. They are saved in the csv data files. Every line of these files consists of an image, i.e. 785 numbers between 0 and 1. The first number of each line is the label, i.e. the digit which is depicted in the image. The following 784 numbers are the pixels of the 28 x 28 image.

# Reading and Organizing Data Values and Labels

```python
mnist = pd.read_csv("mnist.csv")
```

```python
label = mnist.iloc[:, 0]
```

```python
label = label.astype(int)
```

```python
pixels= mnist.iloc[:,1:]
```

```python
print(label.shape, pixels.shape)
```

# Standardizing data
    - Standardize features by removing the mean and scaling to unit variance

```python
scaler = StandardScaler()
scaled_pixels = scaler.fit_transform(pixels)
```

# Performing PCA

    - Perform PCA on the dataset to reduce each sample into a 10-dimensional feature vector.

```python
PCA_Model = PCA(n_components= 10).fit(scaled_pixels)
```

```python
Cov1= PCA_Model.get_covariance()
```

```python
scaled_pixels.shape
```

# Component Visualization

```python
Model_components = pd.DataFrame(PCA_Model.components_,
                            index = ["PCA1", "PCA2","PCA3", "PCA4","PCA5", "PCA6","PCA7", "PCA8","PCA9", "PCA10" ])
```

```python
Model_components
```

```python
plt.figure(figsize= (30,10))
sns.heatmap(Model_components, annot = True)
plt.show()
```

# Data Transformation
       - Show the covariance matrix of the transformed data. Please also copy your code snippet here.

```python
PCA_Pixels_transformed = PCA_Model.transform(scaled_pixels)
PCA_Pixels_transformed
```

```python
Covariance_Matrix = np.cov(PCA_Pixels_transformed.T)
```
